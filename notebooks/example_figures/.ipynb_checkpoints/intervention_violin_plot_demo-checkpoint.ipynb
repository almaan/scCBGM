This notebook reproduces the example figure showing how interventions and their results can be visualized.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Set random seed for reproducibility

```python
np.random.seed(42)
```

Create sampler to generate random data (this is tailored to look like the real results)

```python
def sampler(n_obs: int = 1000, mu: float = 1, d = 1, m = 2, p = 0.2, og = False):

    if og:
        n1 = int(round(n_obs * p))
        n2 = n_obs - n1
    else:
        n1 = n_obs
        n2 = 0
    eps = np.abs(np.random.normal(0,1, size = n1))
    x1 = np.random.gamma(5, 10, size = n1)
    x1 = x1 / x1.mean() + mu
    x2 = np.random.normal( d * mu * m, 1, size = n2)
    x12 = np.concatenate((x1,x2))
    return x12
```

Sample data

```python
n_obs = 1000

x_pos_og = sampler(n_obs, 2.1, d = 1, og = False)
x_pos_gn = sampler(n_obs, 3.5, d = 1)

x_neg_og = sampler(n_obs, 1.9, d = 1, og = False)
x_neg_gn = sampler(n_obs, 0.8, d = -1)
```

Create dataframe


```python

x = np.concatenate((x_pos_og, x_pos_gn, x_neg_og, x_neg_gn))
s = np.array(['positive'] * n_obs + ['positive'] * n_obs + ['negative'] * n_obs + ['negative'] * n_obs)
d = np.array(['generated'] * n_obs + ['perturbed'] * n_obs + ['generated'] * n_obs + ['perturbed'] * n_obs)


data = pd.DataFrame({'Data' : d, 'Gene Expression' : x, 'Gene Set' : s })
```

Get means

```python
means = data.groupby(['Data','Gene Set']).agg('mean')
```

Create plot

```python
fig, ax = plt.subplots(1,1, figsize = (5,4))
sns.violinplot(data,
               x = 'Gene Set',
               y = 'Gene Expression',
               hue = 'Data',
               split=True,
               inner = "quart",
               ax = ax,
               palette = {'generated' : '#faf0ca', 'perturbed' : '#8d99ae' },
               edgecolor = 'black',
               width=0.5,
              )

ax.legend(
    loc='upper right',
    #bbox_to_anchor=(1, 1),
    fontsize = 15,
    # Position the legend outside the plot
)


ax.set_xlabel("Gene Set", fontsize=16)  # Set x-label and its font size
ax.set_ylabel("Gene Expression", fontsize=16)  # Set y-label and its font size

# Change font size of x and y ticks
ax.tick_params(axis='x', labelsize=14, rotation = 0)  # Set x-tick label font siz



p0 = float(means.loc['generated','positive'])
dp =  float(means.loc['perturbed','positive']) - p0

ax.arrow(x = 0.3, y = p0 , dx = 0, dy = dp,
            width=0.03,  # Thin line
            head_width=0.1,  # Width of the triangular head
            head_length=0.3,  # Length of the triangular head
            color='#0cce6b',  # Arrow color
            length_includes_head=True,
            )# Include the head in total arrow length



n0 = float(means.loc['generated','negative'])
dn =  float(means.loc['perturbed','negative']) - n0

ax.arrow(x = 1 + 0.3, y = n0 , dx = 0, dy = dn,
            width=0.03,  # Thin line
            head_width=0.1,  # Width of the triangular head
            head_length=0.3,  # Length of the triangular head
            color='#d62828',  # Arrow color
            length_includes_head=True,
            )# Include the head in total arrow length


ax.set_title(r'Intervention: $0\rightarrow 1$', fontsize = 18)
plt.show()
```
