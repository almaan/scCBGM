```python
import pandas as pd
import conceptlab as clab
```

```python
entity, project = "rb-aiml", "clab_hparam_sweep"
df = clab.utils.wandb.download_wandb_project(project, entity)
```

```python
df.head()
```

```python
df = df[df.model.values == 'cem_vae']
```

```python
coi = ['AVG_ACC', 'test_MSE_loss','beta',
       'hidden_dim', 'latent_dim', 'n_concepts', 'min_bottleneck_size',
       'concepts_hp', 'orthogonality_hp','lr']
```

```python
df_grouped = clab.utils.wandb.collapse_columns(df[coi], collapse_columns=['constants.seed', 'n_concepts','AVG_ACC', 'test_MSE_loss'])
```

```python
top_models = clab.utils.wandb.find_top_models(df_grouped, target_columns=['AVG_ACC','test_MSE_loss'], target_direction=['max','min'])
```

```python
top_models.sort_values('test_MSE_loss')
```

```python
top_models.sort_values('test_MSE_loss').iloc[0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.scatter(df_grouped['test_MSE_loss'], df_grouped['AVG_ACC'])
plt.scatter(top_models['test_MSE_loss'], top_models['AVG_ACC'], c  = 'orange')
plt.ylabel('AVG_ACC')
plt.xlabel('test_MSE_loss')
```
