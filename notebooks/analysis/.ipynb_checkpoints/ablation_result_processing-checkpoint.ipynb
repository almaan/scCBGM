```python
import pandas as pd
import conceptlab as clab
```

```python
df = pd.read_csv('../../results/sweeps/wandb_performance_sweep.csv')
```

Define columns of interest

```python
coi = ['model',
       'dataset.n_concepts',
       'constants.seed',
       'AVG_ACC',
       'test_MSE_loss',
       'test_avg_concept_acc',
       'modify.params.n_mod',
       'modify',
       'model.given_gt',
      ]
```

```python
sub_df = df.loc[:,coi].sort_values(by = 'test_MSE_loss').copy()
```

```python
keep_models = ['cb_vae','cb_vae_w_skip', 'cb_vae_w_ccl','skip_cb_vae']
```

```python
sub_df = sub_df.iloc[np.isin(sub_df['model'].values,keep_models)]
```

```python
sub_df['model'] = sub_df['model'].map({
                                     'cb_vae':'cb_vae',
                                     'skip_cb_vae':'scCBGM',
                                     'cb_vae_w_skip' : 'cb_vae_w_skip',
                                     'cb_vae_w_ccl' : 'cb_vae_w_ccl',
                                    })
```

Configure data frame cateegory order

```python
sub_df['model'] = pd.Categorical(sub_df['model'],
                                 categories= ['cb_vae','cb_vae_w_ccl', 'cb_vae_w_skip','scCBGM'])
```

```python
sub_df['corruption'] = pd.Categorical(sub_df['modify'],
                                      categories= ['default',
                                                   'add',
                                                   'drop',
                                                   'duplicate',
                                                   'noise'])
```

Drop irrelevant columns

```python
sub_df.drop(inplace=True,columns=['modify','model.given_gt'])
```

```python
hash_col = [
       'dataset.n_concepts',
       'constants.seed',
       'corruption',
       'modify.params.n_mod'
      ]
```

Generate tables for paper

```python
base_model = 'scCBGM'
mean_list, std_list = [],[]
for k,agg_col in enumerate(['test_MSE_loss', 'AVG_ACC']):

    mu_df = sub_df[['model',agg_col]].groupby('model').agg('mean')
    std_df = sub_df[['model',agg_col]].groupby('model').agg('std')

    #mu_delta_df = mu_df - mu_df.loc[base_model]
    #std_delta_df = np.sqrt(std_df ** 2 + std_df.loc['cb_vae']**2)
    #mu_delta_df = mu_df / mu_df.loc[base_model]
    mu_delta_df = (mu_df - mu_df.loc[base_model]) / mu_df.loc[base_model]
    std_delta_df = mu_df * np.sqrt( ( (std_df / mu_df) ** 2 + (std_df.loc[base_model] / mu_df.loc[base_model])**2 ))

    # convert to percentage
    mu_delta_df = mu_delta_df *100
    std_delta_df = (std_delta_df) * 100

    mean_list.append(mu_delta_df)
    std_list.append(std_delta_df)
```

```python
mu_df = pd.concat(mean_list, axis =1) 
std_df = pd.concat(std_list, axis =1)
```

```python
def prepend_plus(value):
    if not value.startswith('-'):
        return '\\textcolor{mygreen}{' + f'+{value}' + '}'
    else:
        return '\\textcolor{myred}{' + value + '}'
    return value
    
def harmonize_decimals(value):
    if len(value.split('.')[-1]) < 3:
        return f'{value}0'
    return value
```

```python
model_name_map = {'cb_vae' : 'scCBGM - skip - CCL',
                   'cb_vae_w_ccl' : 'scCBGM - skip',
                   'cb_vae_w_skip' : 'scCBGM - CCL',
                    'scCBGM' : 'scCBGM',
                     }
```

```python
pretty_res = mu_df.round(3).astype(str)
pretty_res = pretty_res.applymap(harmonize_decimals)
pretty_res = pretty_res + '\\%'
pretty_res = pretty_res.applymap(prepend_plus)
pretty_res.drop(index = [base_model], inplace = True)

pretty_res.index = pretty_res.index.map(model_name_map)

pretty_res.index.name = None
pretty_res.columns = ['$\\Delta$Reconstruction ($\\downarrow$)', '$\\Delta$Intervention ($\\uparrow$)']
```

```python
print(pretty_res.to_latex(column_format='l' + 'c' * len(pretty_res.columns)))
```

## Full tables

```python
pretty_results = []
for k,values in enumerate(['test_MSE_loss', 'AVG_ACC']):

    ## per corruption
    mean_vals = sub_df[['model','corruption',values]].groupby(['model','corruption'], as_index=False).agg('mean').pivot(index='corruption', columns = 'model', values = values)
    std_vals = sub_df[['model','corruption',values]].groupby(['model','corruption'], as_index=False).agg('std').pivot(index='corruption', columns = 'model', values = values)
    pretty_res = mean_vals.round(3).astype(str).applymap(harmonize_decimals) + r'$\pm$'  + std_vals.round(3).astype('str').applymap(harmonize_decimals) 

    # overall performance
    mean_total = sub_df[[values,'model']].groupby(['model'], as_index=True).agg('mean')
    mean_total.columns = ['all']
    mean_total = mean_total.T
    
    std_total = sub_df[[values,'model']].groupby(['model'], as_index=True).agg('std')
    std_total.columns = ['all']
    std_total = std_total.T

    pretty_total = mean_total.round(3).astype(str) + r'$\pm$'  + std_total.round(3).astype('str') 


   
    pretty_res = pd.concat((pretty_res, pretty_total), axis = 0)

        
    pretty_res.index.name = None
    pretty_res.columns.name = None
    pretty_results.append(pretty_res.copy())
    
    
    print(f'\n======= Table {k+1} ======\n')
    
    pretty_res.columns = pretty_res.columns.map(model_name_map)
    
    pretty_res.index = pretty_res.index.map({'add':'Add',
                                             'noise':'Noise',
                                             'drop':'Drop',
                                             'duplicate':'Duplicate',
                                             'default':'None',
                                             'all' : 'All',
                                             'delta' : "$\Delta$Corrupt",
                                            })

    

    # Split the LaTeX table into lines
    latex_print = pretty_res.to_latex(column_format='c' * (1 + pretty_res.shape[1])).split("\n")
    
    # Insert a custom horizontal line after the second row (adjust indices as needed)
    latex_print.insert(-4, "\\hline")  # Insert a single horizontal line
    # lines.insert(5, "\\hline\\hline")  # For a double horizontal line
    
    # Join the lines back into a single string
    latex_print = "\n".join(latex_print)
    
    print(latex_print)
```
