```python
%load_ext autoreload
%autoreload 2
```

```python
import pandas as pd
import conceptlab as clab
```

```python
df_root = pd.read_csv('../../results/sweeps/wandb_hparam_sweep.csv')
```

```python
df_root.head()
```

## CB-VAE

```python
df = df_root[df_root.model.values == 'cb_vae'].copy()
```

```python
coi = ['AVG_ACC', 'test_MSE_loss','beta',
       'hidden_dim', 'latent_dim', 'n_concepts',
       'min_bottleneck_size','concepts_hp', 'orthogonality_hp','lr']
```

```python
collapse_columns = ['constants.seed','n_concepts','AVG_ACC','test_MSE_loss']

df_grouped = clab.utils.wandb.collapse_columns(df[coi],
                                               collapse_columns=collapse_columns)
```

```python
top_models = clab.utils.wandb.find_top_models(df_grouped,
                                              target_columns=['AVG_ACC','test_MSE_loss'],
                                              target_direction=['max','min'])
```

```python
top_models.sort_values('test_MSE_loss')
```

Pick model

```python
top_models.sort_values('test_MSE_loss').iloc[1]
```
